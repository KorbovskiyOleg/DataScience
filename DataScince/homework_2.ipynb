```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1,2,3,4,5,6,7]
y = [3.5, 3.8, 4.2, 4.5, 5 , 5.5, 7]
```

```python
plt.plot(x,y, marker="o", linestyle = "-")
plt.xlabel("x", fontsize=14, fontstyle="italic", family="monospace", color="darkred")
plt.ylabel("y", fontsize=14, fontstyle="italic", family="monospace", color="darkgreen")
plt.title("График", fontsize=16, fontweight="bold", family="serif", color="darkblue")
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel("x", fontsize=14, fontstyle="italic", family="monospace", color="darkred")
plt.ylabel("y", fontsize=14, fontstyle="italic", family="monospace", color="darkgreen")
plt.title("Диаграмма рассеяния", fontsize=16, fontweight="bold", family="serif", color="darkblue")
plt.show()
```

С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно. Создайте массив Numpy под названием f, содержащий косинусы элементов массива t. Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета. Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'. Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color="green")
plt.title("График f(t)",fontsize=16, fontweight="bold", family="serif", color="darkblue")
plt.xlabel("Значения t",fontsize=14, fontstyle="italic", family="monospace", color="darkred" )
plt.ylabel("Значения f",fontsize=14, fontstyle="italic", family="monospace", color="darkred")
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.show()
```

### *Задание 3

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3
включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с
названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных
графика в сетке, состоящей из двух строк и двух столбцов. 

В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для
координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем
- y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3

```python
x = np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

ax[0, 0].plot(x, y1)
ax[0, 0].set_title('График y1', fontsize=16, fontweight="bold", family="serif", color="darkblue")
ax[0,0].set_xlim(-5,5)


ax[0, 1].plot(x, y2)
ax[0, 1].set_title('График y2',fontsize=16, fontweight="bold", family="serif", color="darkblue")

ax[1, 0].plot(x, y3)
ax[1, 0].set_title('График y3',fontsize=16, fontweight="bold", family="serif", color="darkblue")

ax[1, 1].plot(x, y4)
ax[1, 1].set_title('График y4',fontsize=16, fontweight="bold", family="serif", color="darkblue")
```

```python
fig.subplots_adjust(hspace=0.3, wspace=0.3)
```

```python
plt.show()
```
