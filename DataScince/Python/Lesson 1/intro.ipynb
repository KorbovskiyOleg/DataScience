```python
from tiobeindexpy import tiobeindexpy as tbpy
top_20 = tbpy.top_20()
top_20.drop(['Change', 'Programming Language'], axis=1, inplace=True)
top_20
```

```python
import matplotlib.pyplot as plt
#%matplotlib inline

top_20 = tbpy.top_20()
fig, ax = plt.subplots(1,1)
fig.dpi = 150
plt.barh(top_20['Programming Language.1'][10::-1],
        top_20['Ratings'].apply(lambda x: float(x.strip('%')))[10::-1],
        color= 'red', edgecolor='black')
plt.title('Популярнось языков по индексу TIOBE')
plt.show()
```
